```python
import pandas as pd
```

```python
data=pd.read_csv('./laptop_price.csv', encoding='windows-1251')
data.head()
```

```python
def col_split(x):    
    return x.split(' ')[0]

data['Cpu_Company'] = data['Cpu'].apply(col_split)
data.head()
```

```python
def col_split(x):    
    return x.split(' ')[0]

data['Memory_Amount'] = data['Memory'].apply(col_split)
data.head()
```
